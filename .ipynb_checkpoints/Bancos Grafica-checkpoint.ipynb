```python
#modules
import os
import csv
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import squarify
import numpy as np

#Open CSV in Dataframe

#Import CSV
csvpath = os.path.join("Datasets", "BancosVf.csv")

# Import the Bancos.csv file as a DataFrame
BancosDB = pd.read_csv(csvpath, encoding="utf-8")
BancosDB.head()
```

```python
#BancosDB['Cant.Transacciones por Celular'].mean()


BancosGraph = BancosDB[BancosDB['Cant.Transacciones por Celular'] >= BancosDB['Cant.Transacciones por Celular'].mean()]
BancosGraph.sort_values(by=['Cant.Transacciones por Celular'],ascending=False)
BancosGraph = BancosGraph.sort_values(by=['Cant.Transacciones por Celular'],ascending=False)
```

```python
#Utilise matplotlib to scale our goal numbers between the min and max, then assign this scale to our values.
norm = matplotlib.colors.Normalize(vmin=min(BancosGraph['Cant.Transacciones por Celular']), vmax=max(BancosGraph['Cant.Transacciones por Celular']))
colors = [matplotlib.cm.Blues(norm(value)) for value in BancosGraph['Cant.Transacciones por Celular']]

#Create our plot and resize it.
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 4.5)

#Use squarify to plot our data, label it and add colours. We add an alpha layer to ensure black labels show through
squarify.plot(label=BancosGraph.State,sizes=BancosGraph['Cant.Transacciones por Celular'], color = colors, alpha=.7)
plt.title("Participacion en Transacciones Electronicas",fontsize=23,fontweight="bold")

#Remove our axes and display the plot
plt.axis('off')
plt.show()
```

```python

 
# Basic chart
#df=pd.DataFrame({'x': BancosDB['Cant. Surcursales'], 'y': BancosDB['Cant.'] })
x=BancosDB['Cant. Surcursales']
y=BancosDB['Cant. ATMs']

plt.plot(x, y, linestyle='none', marker='o')
 
# Annotation
plt.title('Cant. Surcursales vs Cant. ATMs ')
plt.xlabel('Cant. Surcursales')
plt.ylabel('Cant. ATMs')

plt.axvline(750, color='r')
plt.axhline(3000, color='green')

 
# add text annotation
plt.text(1820, 8000, "CDMX", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.text(1400, 4900, "MEX", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.text(1080, 3400, "JAL", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.text(900, 3900, "NLE", horizontalalignment='left', size='medium', color='black', weight='semibold')


```
