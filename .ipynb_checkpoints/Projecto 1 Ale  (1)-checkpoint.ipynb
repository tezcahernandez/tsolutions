```python
#import my dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from CSV
educacion = './educacion.xlsx'
educacion_1 = pd.read_excel(educacion)
educacion_1.head()
```

```python
#Summary of df
educacion_1.describe()
```

```python
#Data types -*shouldnt they be all int
educacion_1.dtypes
```

```python
#DF shape
educacion_1.shape
```

```python
#All states in DF
educacion_1["Entidad federativa"].unique()
```

```python
# Finding how many students each trainer has - como se busca por estado?
educacion_1["Entidad federativa"].sum["Sin escolaridad"]
```

```python
# Calculate the min/max no scolarity per state?????
minsin_escolaridad = educacion_1["Sin escolaridad"].min()
maxsin_escolaridad = educacion_1["Sin escolaridad"].max()


# Place all of the data found into a summary DataFrame
summary_table= pd.DataFrame({"Min Sin escolaridad":[minsin_escolaridad],"Max Sin Escolaridad":[maxsin_escolaridad]})
summary_table
```

```python
#Calculate the total education levels for the entire dataset
total_educacion = educacion_1["Sin escolaridad"].sum() + educacion_1["Educación básica total"].sum(
) + educacion_1["Secundaria"].sum() + educacion_1["Educación media superior2"].sum() + educacion_1["Grado promedio de escolaridad"].sum()

# Place all of the data found into a summary DataFrame
summary_table= pd.DataFrame({"Total":[total_educacion]})
summary_table
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(educacion_1))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, educacion_1["Sin escolaridad"], color='r', alpha=0.5, align="center")
```

```python
plt.xticks(tick_locations, educacion_1["Entidad federativa"], rotation="vertical")
```

```python
# Set a Title and labels
plt.title("No scolarity average per State")
plt.xlabel("State")
plt.ylabel("Average No Scolarity")
```

```python
#CHART DATA FRAME --- No Scolarity Per State 
# Filter the DataFrame down only to those columns to chart
state_and_no_scolarity = educacion_1[["Entidad federativa","Sin escolaridad"]]

# Set the index to be "State" so they will be used as labels
state_and_no_scolarity = state_and_no_scolarity.set_index("Entidad federativa")
state_and_no_scolarity.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
state_and_no_scolarity.plot(kind="bar", figsize=(20,3))

# Set a title for the chart
plt.title("Average No Scolarity Per State")

plt.show()
plt.tight_layout()
```
