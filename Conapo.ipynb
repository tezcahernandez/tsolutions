```python
# Dependencies and Setup
import pandas as pd
import csv
import matplotlib.pyplot as plt


# File to Load (Remember to Change These)
proyecciones_to_load = "ind_dem_proyecciones.csv"

# Read Purchasing File and store into Pandas data frame
proyecciones_to_load = pd.read_csv("ind_dem_proyecciones.csv", encoding='latin-1')
```

```python
proyecciones_to_load.head()
```

```python
proytl2=proyecciones_to_load[['AÑO','ENTIDAD','CRE_TOT','EV','POB_MIT_AÑO','MUJ_15_29','MUJ_30_64','HOM_15_29','HOM_30_64']]
```

```python
proytl2.head()
```

```python
proytl2 = proytl2.loc[proytl2["ENTIDAD"]!="República Mexicana"]
```

```python
proytl2 = proytl2.loc[proytl2["AÑO"]>=2016]
```

```python
proytl2 = proytl2.loc[proytl2["AÑO"]<=2025]
```

```python
proytl2
```

```python
proytl2 ['MUJTOTAL'] = proytl2 ['MUJ_15_29'] + proytl2 ['MUJ_30_64']
```

```python
proytl2 ['HOMTOTAL'] = proytl2 ['HOM_15_29'] + proytl2 ['HOM_30_64']
```

```python
proytl2.head()
```

```python
xd = proytl2.loc[:,["AÑO", "ENTIDAD", "POB_MIT_AÑO"]]
xd = xd.pivot(index='ENTIDAD', columns='AÑO', values='POB_MIT_AÑO').T
xd = xd.reset_index()
xd
```

```python
top8States = proytl2.groupby("ENTIDAD").sum()["POB_MIT_AÑO"].nlargest(8).keys().tolist()
top8States
```

```python
xd.plot(x='AÑO', y='')
```
